```python
import pandas as pd
import numpy as np
import random
import statistics
from sklearn.linear_model import LinearRegression
import time

data = pd.read_excel("C:\\Users\\1roks\\Documents\\FMF\Matematika z računalnikom\\Project\\Data\\Final data\\Final_data.xlsx")
# Izberi nakjlučni vzorec velikosti 300

# Clean sample

data_test = data[data.Train_test_identificator == "test"]
data_train = data[data.Train_test_identificator == "train"]


data_train = data_train.reset_index()
data_test = data_test.reset_index()


group_data = pd.read_excel("C:\\Users\\1roks\\Documents\\FMF\Matematika z računalnikom\\Project\\Data\\Final data\\Group_Data.xlsx")

decili_ccm_kw = pd.read_excel("C:\\Users\\1roks\\Documents\\FMF\Matematika z računalnikom\\Project\\Data\\Final data\\decili_ccm_kw.xlsx")
decili_ccm = list(decili_ccm_kw["decili_ccm"].values)
decili_kw = list(decili_ccm_kw["decili_kw"].values)

def is_nan(n):
    if str(n) == "nan":
        return True
    return False


# Definiraj funkcije za primerjavo

def primerjava_tipa(avto_main_tip, avto_main_znamka, avto_comp_tip, avto_comp_znamka, alfa_tip):
    # Če sta enaka vrni alfa
    if avto_main_tip == avto_comp_tip and avto_main_znamka == avto_comp_znamka:
        return alfa_tip
    
    # Če nista poglej, če imata oba count večji od 3
    
    main_data = group_data[(group_data.tip == avto_main_tip) & (group_data.znamka == avto_main_znamka)]
    
    comp_data = group_data[(group_data.tip == avto_comp_tip) & (group_data.znamka == avto_comp_znamka)]
    
    if len(main_data.index) > 0 and len(comp_data.index) > 0:
        main_mean = main_data["mean"].values[0]
        comp_mean = comp_data["mean"].values[0]
        
        rel_diff = abs((comp_mean - main_mean) / main_mean)
        
        return (1-rel_diff) * alfa_tip
    
    return 0

def primerjava_starosti(main_prva_reg, comp_prva_reg, alfa_starost):
    if is_nan(main_prva_reg) == False and is_nan(comp_prva_reg) == False:
        diff = min(abs(main_prva_reg-comp_prva_reg),30)
        return (30 - diff) * alfa_starost
    return 0

def primerjava_ccm(main_ccm_class, comp_ccm_class, alfa_ccm):
    if is_nan(main_ccm_class) == False and is_nan(comp_ccm_class)== False: 
        diff = 10- abs(main_ccm_class - comp_ccm_class)
        return diff *alfa_ccm
    return 0

def primerjava_kw(main_kw_class, comp_kw_class, alfa_kw):
    if is_nan(main_kw_class) == False and is_nan(comp_kw_class)== False: 
        diff = 10- abs(main_kw_class - comp_kw_class)
        return diff *alfa_kw
    return 0

def primerjava_prevozeni_km(main_km, comp_km, alfa_km):
    if is_nan(main_km) == False and is_nan(comp_km)== False:
        diff = abs(main_km-comp_km)
        diff = min(diff, 300000)
        return alfa_km * (300000 - diff)
    return 0

def primerjava_goriva(main_gorivo, comp_gorivo, alfa_gorivo):
    if is_nan(main_gorivo) == False and is_nan(comp_gorivo)== False:
        if main_gorivo == comp_gorivo:
            return alfa_gorivo
    return 0

def primerjava_menjalnik(main_menjalnik, comp_menjalnik, alfa_menjalnik):
    if is_nan(main_menjalnik) == False and is_nan(comp_menjalnik)== False:
        if main_menjalnik == comp_menjalnik:
            return alfa_menjalnik
    return 0

def primerjava_oblika(main_oblika, comp_oblika, alfa_oblika):
    if is_nan(main_oblika) == False and is_nan(comp_oblika)== False:
        if main_oblika == comp_oblika:
            return alfa_oblika
    return 0
    
    
def primerjava_barva(main_barva, comp_barva, alfa_barva):
    if is_nan(main_barva) == False and is_nan(comp_barva)== False:
        if main_barva == comp_barva:
            return alfa_barva
    return 0


# Definiraj adjustmente

# Starost adjustment


data_for_starost = data_train.dropna(subset=["starost"])

X = data_for_starost["starost"].values
y = data_for_starost["cena"].values

X = X.reshape(-1, 1)
y = y.reshape(-1, 1)

reg_starost = LinearRegression().fit(X, y)


def adjustment_age(starost_main, starost_comp):
        if is_nan(starost_main) == False and is_nan(starost_comp) == False:
            prilagoditev = reg_starost.predict(np.array([[starost_main]])) / reg_starost.predict(np.array([[starost_comp]])) -1
            prilagoditev = prilagoditev[0][0]
            
            return prilagoditev
        else:
            return 0


# Moc motorja v KW

data_for_kw = data_train.dropna(subset = ["kw_class"])

X = data_for_kw["kw_class"].values
y = data_for_kw["cena"].values

X = X.reshape(-1, 1)
y = y.reshape(-1, 1)

reg_kw = LinearRegression().fit(X, y)

  
    
def adjustment_kw(kw_main, kw_comp):
        if is_nan(kw_main) == False and is_nan(kw_comp) == False:
            prilagoditev = reg_kw.predict(np.array([[kw_main]])) / reg_kw.predict(np.array([[kw_comp]])) -1
            prilagoditev = prilagoditev[0][0]
            return prilagoditev
        else:
            return 0
        
        
# Prevozeni km

data_for_prev_km = data_train.dropna(subset = ["prevozeni_km"])

X = data_for_prev_km["prevozeni_km"].values
y = data_for_prev_km["cena"].values

X = X.reshape(-1, 1)
y = y.reshape(-1, 1)

reg_km = LinearRegression().fit(X, y)

def adjustment_prevozeni_km(km_main, km_comp):
        if is_nan(km_main) == False and is_nan(km_comp) == False:
            prilagoditev = reg_km.predict(np.array([[km_main]])) / reg_km.predict(np.array([[km_comp]])) -1
            prilagoditev = prilagoditev[0][0]            
            return prilagoditev
        else:
            return 0
```

```python



def Izracunaj_mere_podobnosti(main_car, comparable_cars,parameters, number_of_neighbours = 5):
    
    # Odstrani glede na prevelka razlika v letih
    
    mere_podobnosti = []
    
    tip = []
    starost = []
    ccm = []
    kw = []
    km = []
    gorivo = []
    menjalnik = []
    oblika = []
    barva = []
    
    alfa_tip = parameters[0]
    alfa_starost = parameters[1]
    alfa_ccm = parameters[2]
    alfa_kw = parameters[3]
    alfa_km = parameters[4] # Rationale 100,000 km je prbližno enako 3 leta starosti
    alfa_gorivo = parameters[5]
    alfa_menjalnik = parameters[6]
    alfa_oblika = parameters[7]
    alfa_barva = parameters[8]

    
    for index, row in comparable_cars.iterrows():
        podobnost = 0
        
        # Primerjava Tipa
        
        avto_main_tip = main_car["tip"].values[0]
        avto_main_znamka = main_car["znamka"].values[0]
        avto_comp_tip = row["tip"]
        avto_comp_znamka = row["znamka"]
        
        p = primerjava_tipa(avto_main_tip, avto_main_znamka, avto_comp_tip, avto_comp_znamka, alfa_tip)
        tip.append(p)
        podobnost += p
        
        # Primerjava starosti
        
        main_prva_reg =  main_car["leto_prve_registracije"].values[0]
        comp_prva_reg =  row["leto_prve_registracije"]
        
        p =  primerjava_starosti(main_prva_reg, comp_prva_reg, alfa_starost)
        starost.append(p)
        podobnost += p     
        
        # Primerjava ccm
        
        main_ccm = main_car["ccm_class"].values[0]
        comp_ccm = row["ccm_class"]
        
        p = primerjava_ccm(main_ccm, comp_ccm, alfa_ccm)
        ccm.append(p)
        podobnost += p
        
        # Primerjava kw
        
        main_kw =  main_car["kw_class"].values[0]
        comp_kw = row["kw_class"]
        
        p = primerjava_kw(main_kw, comp_kw, alfa_kw)
        kw.append(p)
        podobnost += p
        
        # Prevozeni km
        
        main_km =  main_car["prevozeni_km"].values[0]
        comp_km = row["prevozeni_km"]
        
        p = primerjava_prevozeni_km(main_km, comp_km, alfa_km)
        km.append(p)
        podobnost += p
        
        # Primerjava gorivo
        
        main_gorivo =  main_car["gorivo"].values[0]
        comp_gorivo = row["gorivo"]
        
        p = primerjava_goriva(main_gorivo, comp_gorivo, alfa_gorivo)
        gorivo.append(p)
        podobnost += p
        
        
        # Menjalnik
        
        main_menjalnik = main_car["menjalnik"].values[0]
        comp_menjalnik =row["menjalnik"]
        
        p = primerjava_menjalnik(main_menjalnik, comp_menjalnik, alfa_menjalnik)
        menjalnik.append(p)
        podobnost += p
        
        # Oblika
        
        main_oblika= main_car["oblika"].values[0]
        comp_oblika = row["oblika"]
        
        p = primerjava_oblika(main_oblika, comp_oblika, alfa_oblika)
        oblika.append(p)
        podobnost += p
        
        # Barva
        
        main_barva = main_car["barva"].values[0]
        comp_barva = row["barva"]
        
 
        p =  primerjava_barva(main_barva, comp_barva, alfa_barva)
        barva.append(p)
        podobnost += p
        
        # Konec primerjav
        
        mere_podobnosti.append(podobnost)
        
    comparable_cars["mere_podobnosti"] = mere_podobnosti
    comparable_cars["atribut_tip"] = tip 
    comparable_cars["atribut_starost"] =starost
    comparable_cars["atribut_ccm"] = ccm 
    comparable_cars["atribut_kw"] =kw 
    comparable_cars["atribut_prev_km"] =km 
    comparable_cars["atribut_gorivo"] =gorivo 
    comparable_cars["atribut_menjalnik"] =menjalnik 
    comparable_cars["atribut_oblika"] =oblika 
    comparable_cars["atribut_barva"] =barva 
    
    
    comparable_cars = comparable_cars.sort_values(by= "mere_podobnosti", ascending =False)
    
    comparable_cars_final_candidates = comparable_cars.head(number_of_neighbours)
    
    
    return comparable_cars_final_candidates["cena"].mean()

def Izracunaj_mere_podobnosti_s_prilagoditvami(main_car, comparable_cars,parameters, number_of_neighbours = 5):
    
    start = time.time()
    # Odstrani glede na prevelka razlika v letih
    
    mere_podobnosti = []
    
    tip = []
    starost = []
    ccm = []
    kw = []
    km = []
    gorivo = []
    menjalnik = []
    oblika = []
    barva = []
    
    alfa_tip = parameters[0]
    alfa_starost = parameters[1]
    alfa_ccm = parameters[2]
    alfa_kw = parameters[3]
    alfa_km = parameters[4] # Rationale 100,000 km je prbližno enako 3 leta starosti
    alfa_gorivo = parameters[5]
    alfa_menjalnik = parameters[6]
    alfa_oblika = parameters[7]
    alfa_barva = parameters[8]
    
    
    for index, row in comparable_cars.iterrows():
        podobnost = 0
        
        # Primerjava Tipa
        
        avto_main_tip = main_car["tip"].values[0]
        avto_main_znamka = main_car["znamka"].values[0]
        avto_comp_tip = row["tip"]
        avto_comp_znamka = row["znamka"]
        
        p = primerjava_tipa(avto_main_tip, avto_main_znamka, avto_comp_tip, avto_comp_znamka, alfa_tip)
        tip.append(p)
        podobnost += p
        
        # Primerjava starosti
        
        main_prva_reg =  main_car["leto_prve_registracije"].values[0]
        comp_prva_reg =  row["leto_prve_registracije"]
        
        p =  primerjava_starosti(main_prva_reg, comp_prva_reg, alfa_starost)
        starost.append(p)
        podobnost += p     
        
        # Primerjava ccm
        
        main_ccm = main_car["ccm_class"].values[0]
        comp_ccm = row["ccm_class"]
        
        p = primerjava_ccm(main_ccm, comp_ccm, alfa_ccm)
        ccm.append(p)
        podobnost += p
        
        # Primerjava kw
        
        main_kw =  main_car["kw_class"].values[0]
        comp_kw = row["kw_class"]
        
        p = primerjava_kw(main_kw, comp_kw, alfa_kw)
        kw.append(p)
        podobnost += p
        
        # Prevozeni km
        
        main_km =  main_car["prevozeni_km"].values[0]
        comp_km = row["prevozeni_km"]
        
        p = primerjava_prevozeni_km(main_km, comp_km, alfa_km)
        km.append(p)
        podobnost += p
        
        # Primerjava gorivo
        
        main_gorivo =  main_car["gorivo"].values[0]
        comp_gorivo = row["gorivo"]
        
        p = primerjava_goriva(main_gorivo, comp_gorivo, alfa_gorivo)
        gorivo.append(p)
        podobnost += p
        
        
        # Menjalnik
        
        main_menjalnik = main_car["menjalnik"].values[0]
        comp_menjalnik =row["menjalnik"]
        
        p = primerjava_menjalnik(main_menjalnik, comp_menjalnik, alfa_menjalnik)
        menjalnik.append(p)
        podobnost += p
        
        # Oblika
        
        main_oblika= main_car["oblika"].values[0]
        comp_oblika = row["oblika"]
        
        p = primerjava_oblika(main_oblika, comp_oblika, alfa_oblika)
        oblika.append(p)
        podobnost += p
        
        # Barva
        
        main_barva = main_car["barva"].values[0]
        comp_barva = row["barva"]
        
 
        p =  primerjava_barva(main_barva, comp_barva, alfa_barva)
        barva.append(p)
        podobnost += p
        
        # Konec primerjav
        
        mere_podobnosti.append(podobnost)
        
    comparable_cars["mere_podobnosti"] = mere_podobnosti
    comparable_cars["atribut_tip"] = tip 
    comparable_cars["atribut_starost"] =starost
    comparable_cars["atribut_ccm"] = ccm 
    comparable_cars["atribut_kw"] =kw 
    comparable_cars["atribut_prev_km"] =km 
    comparable_cars["atribut_gorivo"] =gorivo 
    comparable_cars["atribut_menjalnik"] =menjalnik 
    comparable_cars["atribut_oblika"] =oblika 
    comparable_cars["atribut_barva"] =barva 
    
    
    comparable_cars = comparable_cars.sort_values(by= "mere_podobnosti", ascending =False)
    
    comparable_cars_final_candidates = comparable_cars.head(number_of_neighbours)
    
    # Adjustmente na teh 5ih
    kocne_prilagoditve = []
    
    for index, row in comparable_cars_final_candidates.iterrows():
        
        # Adjustment Prevozeni km
        
        
        km_main = main_car["prevozeni_km"].values[0]
        km_comp = row["prevozeni_km"]
        
        prilagoditev_km = adjustment_prevozeni_km(km_main, km_comp)
        
        # Adjustment Moc motorja
        
        kw_main = main_car["kw_class"].values[0]
        kw_comp = row["kw_class"]
        
        prilagoditev_kw =  adjustment_kw(kw_main, kw_comp)
        
        # Adjustment Starost
        
        starost_main =  main_car["starost"].values[0]
        starost_comp  =row["starost"]
        
        prilagoditev_starost =  adjustment_age(starost_main, starost_comp) 
        
        
        # Prilagoditve skupaj
        
        koncna_prilagoditev = (1+prilagoditev_km) * (1+prilagoditev_kw) *(1+prilagoditev_starost)
        
        kocne_prilagoditve.append(koncna_prilagoditev)
            
    
    
    # Vzami samo avte, ki imajo prilagoditev med 0.5 in 1.5
    
    stevec_prevelikih_prilagoditev = 0
    cene_avtov = list(comparable_cars_final_candidates["cena"].values)
    
    koncna_ocena = 0
    
    for i in range(0,len(cene_avtov)):
        if kocne_prilagoditve[i] <= 1.5 and kocne_prilagoditve[i] > 0.5:
            koncna_ocena += kocne_prilagoditve[i] * cene_avtov[i]
        else:
            stevec_prevelikih_prilagoditev += 1
            
    # Če je avtov s prevelikimi adjustmenti preveč (preostane manj kot trije avti) potem autputej samo povprečje cen
    
    finish = time.time()
    
    print(f"Potreboval {round(finish-start,2)} sekunde")
    
    if len(cene_avtov) - stevec_prevelikih_prilagoditev < 3:
 # return comparable_cars_final_candidates
       return comparable_cars_final_candidates["cena"].mean()
    
    
    return koncna_ocena / (len(cene_avtov) - stevec_prevelikih_prilagoditev)

```

```python
alfa_tip = 125
alfa_starost = 3 
alfa_ccm = 5
alfa_kw = 5
alfa_km = 1 / 3000 # Rationale 100,000 km je prbližno enako 3 leta starosti
alfa_gorivo = 20
alfa_menjalnik = 15
alfa_oblika = 15
alfa_barva = 1

parametri = [alfa_tip,alfa_starost,alfa_ccm,alfa_kw,alfa_km,alfa_gorivo,alfa_menjalnik,alfa_oblika,alfa_barva]

save_all_paramter_config = [parametri]

stevec = 0

train_ids = list(data_test["ID"].values)

current_best_parameters = 0

current_best_median_result = 10 ** 10

save_results = []

for k in range(0,20):
    stevec2 = 0
    rel_diff_all = []
    
    if k > 0:
        
        # Naključno spremeni 3 vrednosti parametrov
        
        for rand in range(0,3):
                
            random_int = random.randrange(0,9)
            random_float = random.random() * 2

            parametri[random_int] *= random_float
            
        print("NOVI PARAMETRI")
        print(parametri)
        
        save_all_paramter_config.append(parametri)
        
    
    for id in train_ids:
        
        print(stevec2)
        stevec2 += 1

        # Oceni vse transakcije v train setu za specifične parametre
        
        main_car = data_test[data_test.ID == id]

        estimate = Izracunaj_mere_podobnosti(main_car, data_train ,parametri, number_of_neighbours = 5)
        
        price= main_car["cena"].values[0]

        abs_diff = abs(estimate- price)

        rel_diff = abs_diff / price

        rel_diff_all.append(rel_diff)
        
        print(f"rel diff is {rel_diff}")
    # Shrani rezultate
    
    save_results.append(rel_diff_all)
    
    # Poglej ali se je napoved izboljšala in če se je jo določi za novo najboljšo
    
    median_abs_rel_error = statistics.median(rel_diff_all)
    
    # Če se je izboljšala nastavi novo najboljšo napako
    
    if median_abs_rel_error <= current_best_median_result:
        current_best_parameters = k
        current_best_median_result = median_abs_rel_error
        
    # Če se ni pojdi nazaj na prejšno konfiguracijo paramtetrov
    
    else:
        parametri = save_all_paramter_config[k-1]
        
    
    
        
            
```
